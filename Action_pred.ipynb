        <h3>2. Marco teórico</h3>
        <h3>Modelo Black-Scholes</h3>
            <h3>&emsp; El modelo Black-Scholes es un método utilizado para calcular el precio de opciones europeas. Este modelo asume que el precio del activo subyacente sigue un proceso de tipo Geometría Browniana, es decir, su evolución es aleatoria pero predecible bajo ciertas condiciones. Además, supone que:</h3>
<ul><h3>
    <li>El precio del activo subyacente sigue un proceso estocástico de tipo <b>Geometría Browniana.</b></li>
    <li>No existen costos de transacción ni impuestos.</li>
    <li>El mercado financiero es eficiente y no permite arbitraje.</li>
    <li>La tasa de interés libre de riesgo es constante y conocida.</li>
    <li>El precio del activo subyacente no paga dividendos durante la vida de la opción.</li>
    <li>La volatilidad del activo subyacente es constante.</h3>
</li>
</h3></ul>
</font>            


<font face="times new roman" size="3" color="black">
    <div align="justify">
        <h3><i>2.1 Fórmula del modelo Black-Scholes para la opción de compra (call):</i></h3>
    </div>
<h2>
$$
\mathbf{C = S_0N(d_1) - N(d_2)Ke^{-rT}}
$$
</h2>
<h3>donde: </h3>

<ul><h3>
    <li><strong>C</strong>: el precio de la opción de compra.</li>
    <li><strong>S₀</strong>: el precio actual del activo subyacente.</li>
    <li><strong>K</strong>: el precio de ejercicio de la opción.</li>
    <li><strong>r</strong>: la tasa de interés libre de riesgo.</li>
    <li><strong>T</strong>: el tiempo hasta el vencimiento (en años).</li>
    <li><strong>N(d)</strong>: la función de distribución acumulativa de una variable normal estándar.</li>
</h3></ul>

<h3>d₁ y d₂ se definen como:</h3>
<h3>
$$
\mathbf{d_1=\frac{\ln(S_0 / K) + (r + \frac{\sigma^2}{2})T}{\sigma \sqrt{T}}},
$$
$$
\mathbf{d_2=d_1 - \sigma \sqrt{T}}
$$
</h3>
<ul><h3>
    <li><strong>σ</strong> es la volatilidad del activo subyacente, que mide la variabilidad de su precio.</li>
</h3></ul>
</font>

<font face="times new roman" size="3" color="black">
    <div align="justify">
        <h3><i>2.2 Fórmula del modelo Black-Scholes para una opción de venta (put):</i></h3>
    </div>
<h2>
$$
\mathbf{P = Ke^{-rT} N(-d_2) - S_0 N(-d_1)}
$$
</h2>
<h3>donde: </h3>
<ul><h3>
    <li><strong>P</strong>: Es el precio asociado de la opción de venta.</li>
    <li>El resto de variables son las mismas que el modelo de compra.</li>
</h3></ul>
    <div align="justify">
<h3>Es importante aclarar ciertas ambigüedades que surgen al desarrollar estás fórmulas. Notamos que:</h3>
    </div>
<ul><h3>
    <li><strong>S₀N(d₁)</strong>: Es el valor esperado de recibir el activo subyacente en el vencimiento, ajustado por la probabilidad de que el precio del activo supere el precio de ejercicio (en el caso call).</li>
    <li><strong>\(\mathbf{Ke^{-rT}}\)N(d₂)</strong>:Es el valor presente del pago del precio de ejercicio, ajustado por la probabilidad de que la opción termine en dinero.
</h3></ul>
</font>

<font face="times new roman" size="3" color="black">
    <div align="justify">
        <h3>Movimiento Browniano</h3>
            <h3>&emsp; El modelo Black-Scholes se deriva bajo la premisa de que el precio del activo subyacente sigue un proceso de Itô, específicamente una geometría Browniana. El movimiento browniano se utiliza para modelar la evolución aleatoria de los precios de los activos, debido a que es capaz de capturar la naturaleza impredecible y volátil de los mercados financieros. En particular, el modelo Black-Scholes asume que el precio de un activo sigue un movimiento browniano geométrico, lo que significa que los precios crecen de manera exponencial bajo un comportamiento estocástico, cuya dinámica se describe por las siguiente ecuación diferencial estocástica: </h3>
<h2>
$$
\mathbf{dS_t = S_t \mu dt + S_t \sigma dW_t}
$$
</h2>
<h3>donde: </h3>
<ul><h3>
    <li><strong>\(\mathbf{S_t}\)</strong>: Es el precio del activo subyacente en el tiempo t.</li>
    <li><strong>μ</strong>: Es la tasa de rendimiento esperada del activo.</li>
    <li><strong>σ</strong>: Es la volatilidad del activo.</li>
    <li><strong>\(\mathbf{W_t}\)</strong>: Es un movimiento browniano estándar.</li>
</h3></ul>
<h3>Se usa una técnica conocida como "cobertura libre de riesgo" (replicación de portafolios) y el teorema de Feynman-Kac para derivar la ecuación de Black-Scholes. La ecuación diferencial parcial que satisface el precio de una opción \(\mathbf{V(S,t)}\) es :</h3>
<h3>
$$
\mathbf{\frac{\partial V}{\partial t} + \frac{1}{2} \sigma^2 S^2 \frac{\partial^2 V}{\partial S^2} + rS \frac{\partial V}{\partial S} - rV = 0}
$$
</h3>

<h3>Método ARCH con cadenas de Markov</h3>

<h3>&emsp;El método ARCH (Autoregressive Conditional Heteroskedasticity), desarrollado por Robert Engle en 1982, es ampliamente utilizado para modelar series temporales con volatilidad heterocedástica, es decir, donde la varianza de los errores varía a lo largo del tiempo. Sin embargo, cuando se combina con cadenas de Markov, se obtiene un modelo más sofisticado conocido como ARCH con cambio de régimen o ARCH-M (Markov-Switching ARCH).</h3>

<h3>En este modelo, se introduce un enfoque de múltiples regímenes, donde la estructura de la volatilidad no es constante, sino que cambia entre diferentes estados de volatilidad (regímenes) siguiendo un proceso de cadenas de Markov. Cada régimen tiene su propia dinámica para la volatilidad condicional, lo que permite capturar mejor los cambios estructurales en los mercados financieros, como períodos de calma seguidos por episodios de alta inestabilidad.</h3>

<h3>La cadenas de Markov describe cómo el sistema cambia de un régimen a otro de manera probabilística, dependiendo solo del régimen actual, sin necesidad de recordar los regímenes anteriores. Este modelo es útil en finanzas porque los mercados suelen presentar periodos prolongados de baja volatilidad (mercados estables) seguidos de cambios abruptos a alta volatilidad (mercados de crisis), y este comportamiento puede ser capturado de forma más precisa mediante el ARCH-M. El modelo ARCH con cadenas de Markov permite modelar mejor la dinámica de la volatilidad en escenarios donde los cambios en el mercado no son constantes, sino que dependen de diferentes estados o condiciones del entorno financiero.</h3>

<h3>Modelo de Montecarlo</h3>

<h3>&emsp;El método de Monte Carlo es una técnica de simulación estocástica que se utiliza para evaluar modelos complejos que no se pueden resolver analíticamente. Se basa en la generación de un gran número de muestras aleatorias para simular diferentes escenarios posibles y obtener estimaciones estadísticas de los resultados. En el contexto financiero, el método de Monte Carlo es ampliamente empleado para la valoración de derivados, la gestión de riesgos y la simulación de trayectorias de precios de activos.</h3>

<h3>Este método es especialmente útil para modelar la incertidumbre y la volatilidad de los mercados financieros, ya que permite generar múltiples trayectorias posibles de un activo subyacente, considerando diferentes variables como el precio inicial, la volatilidad, la tasa de interés libre de riesgo y el tiempo hasta el vencimiento. Estas trayectorias simuladas permiten calcular el valor esperado de una opción, un portafolio o cualquier otro activo financiero.</h3>

<h3>El método de Monte Carlo es esencial en situaciones donde la evolución de los precios sigue procesos estocásticos complejos, como el previamente descrito movimiento browniano geométrico, y cuando las técnicas analíticas tradicionales, como el modelo Black-Scholes, son insuficientes para capturar todas las incertidumbres del mercado.</h3>

<h3>Opciones financieras</h3>

<h3>&emsp;Las opciones financieras son contratos que otorgan el derecho, pero no la obligación, de comprar o vender un activo subyacente a un precio determinado, conocido como el precio de ejercicio, antes o en una fecha específica. El activo subyacente puede ser un valor, un índice, o una mercancía, y el contrato refleja un acuerdo sobre este activo.</h3>

<h3>El precio de ejercicio o "strike price" es el valor al que el titular de la opción puede ejercer su derecho de compra o venta. La diferencia entre las opciones europeas y americanas radica en el momento en que se pueden ejercer. Las opciones europeas sólo se ejercen al vencimiento, mientras que las americanas pueden ejercerse en cualquier momento antes de su vencimiento.</h3>

<h3>En cuanto a los tipos de opciones, la opción de venta (put) otorga el derecho a vender el activo subyacente a un precio fijo, mientras que la opción de compra (call) otorga el derecho a comprarlo.</h3>

<h3>El precio de la prima en una opción europea es el costo que paga el comprador para adquirir la opción. Si el titular de la opción decide no ejercerla en la fecha de vencimiento, el contrato simplemente expira y no se lleva a cabo ninguna transacción.</h3>

<h3>Por ejemplo, si un inversor tiene una opción de compra europea sobre acciones de Tesla con un precio de ejercicio de \$50 y en la fecha de vencimiento el precio de las acciones es inferior a \$50, el inversor probablemente no ejercería la opción, ya que no sería rentable comprar a un precio superior al del mercado. Si el precio de las acciones es superior a \$50, el inversor ejercería su opción para obtener una ganancia, comprando las acciones a un precio menor.</h3>

<h3>En otro caso, si un inversor posee una opción de venta europea sobre acciones de Nvidia con un precio de ejercicio de \$75, y en la fecha de vencimiento el precio de las acciones es superior a ese valor, el inversor no ejercerá su derecho de venta, ya que sería más rentable vender las acciones en el mercado. Si el precio de las acciones es inferior a \$75, el inversor ejercería la opción para vender a un precio mayor al de mercado.</h3>

<h3>El dividendo es una parte de los beneficios de una empresa que se distribuye entre los accionistas, mientras que la volatilidad mide la magnitud de las variaciones del precio del activo subyacente a lo largo del tiempo. Esta volatilidad es clave en el modelo de Black-Scholes, que asume varias condiciones, entre ellas que no se distribuyen dividendos y que los mercados son eficientes.</h3>

<h3>Las limitaciones del modelo Black-Scholes incluyen su incapacidad para manejar opciones americanas, suponer que la volatilidad es constante y que no hay pagos de dividendos. Sin embargo, las variables clave de este modelo incluyen el precio del activo subyacente, el precio de ejercicio, el tiempo hasta el vencimiento, la volatilidad del activo, la tasa de interés libre de riesgo, y los dividendos.</h3>

<h3>La volatilidad implícita en el contexto de Black-Scholes refleja la percepción del mercado sobre el riesgo asociado al precio del activo subyacente. En la fórmula de Black-Scholes, el factor exponencial \(\mathbf{e^{rT}}\) ajusta el valor presente del precio de ejercicio. Un aumento en la volatilidad del activo subyacente eleva el precio de la opción, ya que implica mayores fluctuaciones y, por tanto, más oportunidades de obtener beneficios.</h3>

<h3>Finalmente, un aumento en la tasa de interés libre de riesgo incrementa el precio de las opciones de compra (call) y disminuye el de las opciones de venta (put). Asimismo, una disminución en el tiempo hasta el vencimiento reduce el precio de la opción, ya que hay menos tiempo para que ocurra un movimiento favorable en el precio del activo subyacente.</h3>

# 2. Suponer que los rendimientos pueden ser descritos a través de un movimiento Browniano geométrico

Un proceso estocástico sigue un movimiento Browniano geométrico si sigue

$$
\mathbf{dS_t = S_t \mu dt + S_t \sigma dW_t}
$$

La solución de esta ecuación está dada por:

$$
S_t=S_0\,e^{(\mu - \sigma^2/2)\,t+\sigma\,W_t}
$$

donde:


$\mathbf{S_t}$: Es el precio del activo subyacente en el tiempo t.

$μ$: Es la tasa de rendimiento esperada del activo.

$σ$: Es la volatilidad del activo.

$\mathbf{W_t}$: Es un movimiento browniano estándar.

```python
import random

def GBM(n_steps, mu, sigma,
        t=1, start=0):

    steps = [(mu - (sigma**2)/2)*t + np.random.randn()*sigma for i in range(n_steps)]
    y = start * np.exp(np.cumsum(steps))
    x = [t * i for i in range(n_steps)]

    return x, y

def GBM_2(n_steps, mu, sigma,
        t=1, start=0):

    dt = t / n_steps
    steps = [(mu - 0.5 * sigma**2) * dt + np.random.randn() * sigma * np.sqrt(dt) for i in range(n_steps)]
    y = start * np.exp(np.cumsum(steps))
    x = [dt * i for i in range(n_steps)]
    return x, y
```

<font face="times new roman" size="3" color="black">
    <div align="justify">
        <h3><i>2.1 Fórmula del modelo Black-Scholes para la opción de compra (call):</i></h3>
    </div>
<h2>
$$
\mathbf{C = S_0N(d_1) - N(d_2)Ke^{-rT}}
$$
</h2>
<h3>donde: </h3>

<ul><h3>
    <li><strong>C</strong>: el precio de la opción de compra.</li>
    <li><strong>S₀</strong>: el precio actual del activo subyacente.</li>
    <li><strong>K</strong>: el precio de ejercicio de la opción.</li>
    <li><strong>r</strong>: la tasa de interés libre de riesgo.</li>
    <li><strong>T</strong>: el tiempo hasta el vencimiento (en años).</li>
    <li><strong>N(d)</strong>: la función de distribución acumulativa de una variable normal estándar.</li>
</h3></ul>

<h3>d₁ y d₂ se definen como:</h3>
<h3>
$$
\mathbf{d_1=\frac{\ln(S_0 / K) + (r + \frac{\sigma^2}{2})T}{\sigma \sqrt{T}}},
$$
$$
\mathbf{d_2=d_1 - \sigma \sqrt{T}}
$$
</h3>
<ul><h3>
    <li><strong>σ</strong> es la volatilidad del activo subyacente, que mide la variabilidad de su precio.</li>
</h3></ul>
</font>

<font face="times new roman" size="3" color="black">
    <div align="justify">
        <h3><i>2.2 Fórmula del modelo Black-Scholes para una opción de venta (put):</i></h3>
    </div>
<h2>
$$
\mathbf{P = Ke^{-rT} N(-d_2) - S_0 N(-d_1)}
$$
</h2>
<h3>donde: </h3>
<ul><h3>
    <li><strong>P</strong>: Es el precio asociado de la opción de venta.</li>
    <li>El resto de variables son las mismas que el modelo de compra.</li>
</h3></ul>
    <div align="justify">
<h3>Es importante aclarar ciertas ambigüedades que surgen al desarrollar estás fórmulas. Notamos que:</h3>
    </div>
<ul><h3>
    <li><strong>S₀N(d₁)</strong>: Es el valor esperado de recibir el activo subyacente en el vencimiento, ajustado por la probabilidad de que el precio del activo supere el precio de ejercicio (en el caso call).</li>
    <li><strong>\(\mathbf{Ke^{-rT}}\)N(d₂)</strong>:Es el valor presente del pago del precio de ejercicio, ajustado por la probabilidad de que la opción termine en dinero.
</h3></ul>
</font>

```python
from scipy.stats import norm

def black_scholes(
        S: float,     # Precio actual
        K: float,     # Precio de ejercicio
        T: int,       # Lapso de tiempo (en años)
        r: float,     # Tasa de interés libre de riesgo
        sigma: float, # Volatilidad
        option_type='call'
    ):

    d1 = (np.log(S / K) + (r + 0.5 * sigma ** 2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)

    if option_type == 'call':
        price = S * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)
    elif option_type == 'put':
        price = K * np.exp(-r * T) * norm.cdf(-d2) - S * norm.cdf(-d1)
    else:
        raise ValueError("Opción no válida. Debe ser 'call' o 'put'.")

    return price
```

# 4. Utilizar el método Arch de cadena de Markov para hacer simulaciones de los rendimientos.



El modelo ARCH fue introducido por Robert Engle en 1982 y es útil para modelar series temporales con volatilidad heterocedástica, es decir, donde la variabilidad o dispersión de los datos cambia a lo largo del tiempo. Es particularmente útil en finanzas, donde se observa que los rendimientos de los activos suelen tener periodos de alta volatilidad seguidos de periodos de baja volatilidad. <br>
La ecuación general del modelo ARCH puede ser expresada como:

$\sigma^{2}_{t} = \omega + \alpha_{1}\epsilon^{2}_{t-1}+\alpha_{2}\epsilon^{2}_{t-2}+...+\alpha_{n}\epsilon^{2}_{t-n}$

Donde:

* $\sigma^{2}_{t}$ es la varianza condicional en el tiempo $t_{n}$
* $\omega$ es un parametro constante que captura el nivel base de volatilidad
* $\alpha_{i}$ son los parametros que capturan la dependencia entre los choques previos $\epsilon_{i}$
* $\epsilon_{t}$ es el termino de error o residuo (choque)  en el tiempo $t$

El propósito de un modelo ARCH es capturar la dependencia de la varianza condicional con respecto a los choques previos, de modo que se pueda modelar el comportamiento de la volatilidad a lo largo del tiempo.

se ajusta el modelo ARCH a la columna de rendimientos ('Returns'). El parámetro `vol='ARCH'` indica que estamos utilizando un modelo ARCH puro (sin otros componentes de volatilidad), y `p=1` indica que solo queremos usar un rezago para capturar la volatilidad. Es decir, la volatilidad actual dependerá del cuadrado del residuo en el paso anterior.



```python
from arch import arch_model
# Ajustar el modelo ARCH a los rendimientos
model = arch_model(data['Returns'], vol='ARCH', p=1)
model_fit = model.fit(disp="off")

# Obtener los parámetros del modelo ajustado
omega = model_fit.params['omega']
alpha = model_fit.params['alpha[1]']
residuals = model_fit.resid

# Número de simulaciones y horizonte de tiempo
simulations = 1000
N = len(data['Returns'])  # Usamos el número de datos disponibles para el horizonte de simulación

# Inicializamos una matriz para almacenar los rendimientos simulados
simulated_returns = np.zeros((simulations, N))

# Generar innovaciones (ruido blanco)
np.random.seed(42)
innovations = np.random.normal(size=(simulations, N))
```

Se inicializan los arreglos `return_sim` (para almacenar los rendimientos) y `sigma_sim` (para almacenar las volatilidades condicionales)
<br>
Se calcula el valor inicial de volatilidad $\sigma_{0}$ como $\sqrt{\frac{\omega}{(1-\alpha)}}$
<br>
por ultimo para cada paso $t$
* se calcula la volatilidad condicional $\sigma_{t}$ en funcion de $\omega , \alpha$ y el residuo del periodo anterior.
* se genera el rendimiento en $t$ multiplicando $\sigma_{t}$ por el choque generado aleatoriamente

```python
# Simular rendimientos usando el modelo ARCH
for i in range(simulations):
    returns_sim = np.zeros(N)
    sigma_sim = np.zeros(N)
    sigma_sim[0] = np.sqrt(omega / (1 - alpha))  # Usamos la media condicional de la volatilidad

    for t in range(1, N):
        sigma_sim[t] = np.sqrt(omega + alpha * residuals[t - 1] ** 2)
        returns_sim[t] = sigma_sim[t] * innovations[i, t]

    simulated_returns[i] = returns_sim
```

Después de obtener los rendimientos simulados, se calculan los precios futuros utilizando la relación entre precios y rendimientos. El precio se obtiene como<br> $S_{0}\cdot e^{cumsum(simulatedreturns)}$ <br> donde $S_{0}$ es el precio inicial (precio actual de TSLA), y `cumsum` acumula los rendimientos simulados para generar precios en el tiempo.

```python
simulated_arch_prices = S0 * np.exp(np.cumsum(simulated_returns, axis=1))
```

<h3>El método ARCH es particularmente útil en entornos donde la volatilidad cambia de manera abrupta, como en periodos de crisis financiera o de alta especulación en el mercado. Al modelar la volatilidad condicional, el modelo ARCH captura mejor la realidad de los mercados financieros, proporcionando estimaciones más robustas para la valoración de las opciones.</h3>

<h3>Sin embargo, es importante considerar que tanto el método de Monte Carlo como el modelo ARCH requieren un mayor coste computacional y un conocimiento más profundo de los procesos estocásticos involucrados. En contraste, el modelo de Black-Scholes, aunque podria ser menos preciso en ciertos contextos, es más accesible y fácil de implementar, lo que puede ser ventajoso en situaciones donde se requiere una evaluación rápida o donde el activo subyacente presenta baja volatilidad.</h3>

        <h3>6. Conclusiones</h3>
            <h3>&emsp;<font color = ''>

<h3>En la actividad se demostro que, si bien el modelo de Black-Scholes es una herramienta útil para la valoración de opciones en mercados estables, presenta limitaciones importantes en entornos con alta volatilidad, como es el caso del activo de Tesla. Los métodos de simulación, como Monte Carlo y ARCH, proporcionan estimaciones más realistas al incorporar la incertidumbre y la variabilidad de los rendimientos del activo subyacente.</h3>

<h3>El método de Monte Carlo, al generar diversas trayectorias posibles para los precios futuros del activo, permite evaluar una gama más amplia de escenarios posibles, lo que resulta en una valoración más precisa de las opciones. Por su parte, el modelo ARCH con cadenas de Markov es particularmente útil para modelar la volatilidad condicional, lo que lo convierte en una herramienta valiosa en mercados financieros poco estables.</h3>

<h3>En resumen, la elección del modelo adecuado para la valoración de opciones debe basarse en las características del activo subyacente y las condiciones del mercado. En activos con alta volatilidad y cambios bruscos en el mercado, los métodos de simulación estocástica ofrecen mejores resultados, mientras que el modelo de Black-Scholes puede ser más adecuado en escenarios de menor incertidumbre.</h3>
